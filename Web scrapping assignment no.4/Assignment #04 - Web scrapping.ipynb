<h2><font color="red">Q1. Scrape the details of most viewed videos on YouTube from Wikipedia.
    
Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos
    
You need to find following details:
    
A) Rank
B) Name
C) Artist
D) Upload date
E) Views
</font></h2>

```python
#connecting to google chrome web driver
driver = webdriver.Chrome(r'chromedriver.exe')


# getting the webpage
driver.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')


#waiting to load the web page
time.sleep(2)

#creating empty list to store the data
rank = []
name = []
artist = []
uplaod_date = []
views = []

#to get rank
try:
    ranks=driver.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[1]")
    for i in ranks:  
        rank.append(i.text)
        
except NoSuchElementException:
    rank.append("-")
except StaleElementReferenceException:
    rank.append("-")
    
time.sleep(3)    
#to get name
try:
    names=driver.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[2]")
    for i in names:
        name.append(i.text.split(sep='[')[0].replace('"',''))
        
except NoSuchElementException:
    name.append("-")
except StaleElementReferenceException:
    name.append("-")
    
time.sleep(3)   
#to get artist
try:
    artists=driver.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[3]")
    for i in artists:
        artist.append(i.text)
        
except NoSuchElementException:
    artist.append("-")
except StaleElementReferenceException:
    artist.append("-")
    
    
time.sleep(3)   
#to get upload date
try:
    upload_dates=driver.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[5]")
    for i in upload_dates:
        uplaod_date.append(i.text)
        
except NoSuchElementException:
    uplaod_date.append("-")
except StaleElementReferenceException:
    uplaod_date.append("-")
    
    
time.sleep(3)    
#to get total views
try:
    view=driver.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[4]")
    for i in view:
        views.append(i.text)
        
except NoSuchElementException:
    views.append("-")
except StaleElementReferenceException:
    views.append("-")
    
    
    
driver.close()

# creating dataframe to store & visualize the data in a tabular format
PROCUCTS = pd.DataFrame(list(zip(rank, name, artist, uplaod_date, views)),
                              columns =['Rank', 'Name', 'Artist', 'Upload Date', 'Views'])


# displaying the Dataframe
PROCUCTS
```

<h2><font color="red"> Q5. Scrape the details of trending repositories on Github.com.
    
Url = https://github.com/
    
You have to find the following details:
    
A) Repository title
B) Repository description
C) Contributors count
D) Language used
    
Note: - From the home page you have to click on the trending option from Explore menu through code.
</font></h2>

```python
#creating empty list to store the data
title = []
desc = []
count = []
lang = []
url = []


#connecting to google chrome web driver
driver = webdriver.Chrome(r'chromedriver.exe')


# getting the webpage
driver.get('https://github.com/')
time.sleep(2)
driver.find_element(By.XPATH,'/html/body/div[1]/header/div/div[2]/div/nav/ul/li[3]/button').click()
time.sleep(2)
driver.find_element(By.XPATH,'/html/body/div[1]/header/div/div[2]/div/nav/ul/li[3]/div/ul[3]/li[3]/a').click()
time.sleep(2)

try:
    title_tags =driver.find_elements(By.XPATH,'//h1[@class="h3 lh-condensed"]')
    for i in title_tags:
        title.append(i.text)
except NoSuchElementException:
    title.append('-')
except StaleElementReferenceException:
    title.append('-')


try:
    urls = driver.find_elements(By.XPATH,'//article[@class="Box-row"]/h1/a')
    for i in urls:
        url.append(i.get_attribute('href'))
except NoSuchElementException:
    url.append('-')
except StaleElementReferenceException:
    url.append('-')
    
    
for i in url:
    driver.get(i)
    time.sleep(1)
    try:
        desc_tags = driver.find_element(By.XPATH,'/html/body/div[4]/div/main/turbo-frame/div/div/div/div[3]/div[2]/div/div[1]/div/p')
        desc.append(desc_tags.text)
    except NoSuchElementException:
        desc.append('-')
    except StaleElementReferenceException:
        desc.append('-')
    
    time.sleep(1)
    try:
        count_tags = driver.find_element(By.XPATH, "//h2[@class='h4 mb-3']/a[contains(text(),'Contributors')]/span")
        count.append(count_tags.text)
    except NoSuchElementException:
        count.append('-')
    except StaleElementReferenceException:
        count.append('-')
    
    time.sleep(1)
    all_lang =[]
    try:
        language_tag = driver.find_elements(By.XPATH,"//li[@class='d-inline']//a//span[1]")
        if language_tag:
            for j in language_tag:
                all_lang.append(j.text)
        else:
            all_lang.append('-')
        lang.append(all_lang)
    except NoSuchElementException:
        lang.append('-')
    except StaleElementReferenceException:
        lang.append('-')
        
        
driver.close()


# creating dataframe to store & visualize the data in a tabular format
git_repo = pd.DataFrame({'TITLE':title,'DESCRIPTION':desc,'COUNT':count,'Languges':lang})

git_repo
```
