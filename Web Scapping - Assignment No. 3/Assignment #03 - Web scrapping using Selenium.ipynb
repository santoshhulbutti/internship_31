<h2><font color="red"> Q2: In the above question, now scrape the following details of each product listed in first 3 pages of your search results and save it in a data frame and csv. In case if any product has less than 3 pages in search results then scrape all the products available under that product name. Details to be scraped are:


- "Brand Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability" and “Product URL”.
    
In case, if any of the details are missing for any of the product then replace it by “-“.</font></h2>

```python
#creating empty list to store the data
brand_name = []
product_name = []
price = []
ret = []
exp_del = []
avail = []
prod_url = []


#dummy variable initialization for page count
first_page = 1 #starting page
last_page = 4 # page number to end scaping


#to get all products link from first 3 result pages & appending it into prod_url list
for page in range(first_page,last_page):
    
    url = driver.find_elements_by_xpath('//a[@class="a-link-normal s-no-outline"]')
    x = len(prod_url)
    for i in url:
        prod_url.append(i.get_attribute('href'))
    time.sleep(2)
    
    # locating next page button and clicking
    next_page = driver.find_element_by_xpath('//a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]')
    print(f'results on page number #{page} :', len(prod_url) - x)
    
    if page <last_page-1:
        next_page.click()
        time.sleep(3)
    
print('Total results gathered :', len(prod_url))
print('\n')
print('\n')


#k = 0
# Extracting all required details
for i in prod_url:
    driver.get(i)
    time.sleep(2)
    
    # Extracting Brand Name
    try:
        brand = driver.find_element_by_xpath('//a[@id="bylineInfo"]').text
        
        if 'Visit the ' in brand:
            brand = brand.replace('Visit the ', '').replace(' Store', '')
            brand_name.append(brand)
        
        elif 'Brand: ' in brand:
            brand = brand.replace('Brand: ', '')
            brand_name.append(brand)
 
    except NoSuchElementException:
        brand_name.append('-')
    
#     print(brand_name[k])
    
    # Extracting product name
    try:
        product = driver.find_element_by_xpath('//span[@id="productTitle"]').text
        product_name.append(product)
    except NoSuchElementException:
        product_name.append('-')
    
#     print(product_name[k])
    
    # Extracting Price
    try:
        pricing = driver.find_element_by_xpath('//div[@id="corePrice_feature_div"]/div/span').text
        
        price.append(pricing.replace('\n','.'))
    except NoSuchElementException:
        price.append('-')
    
#     print(price[k])
    
    # Extracting Return or exchange
    try:
        replacement = driver.find_element_by_xpath('//*[@id="RETURNS_POLICY"]/span/div[2]/a')
        ret.append(replacement.text)
    except NoSuchElementException:
        ret.append('-')
    
#     print(ret[k])
    
    
    # Extracting Expected Delivery
    try:
        delivery = driver.find_element_by_xpath('//div[@id="mir-layout-DELIVERY_BLOCK-slot-PRIMARY_DELIVERY_MESSAGE_LARGE"]/span/span')
        exp_del.append(delivery.text)
    except NoSuchElementException:
        exp_del.append('-')
    
#     print(exp_del[k])
    
    # Extracting Availability
    try:
        availability = driver.find_element_by_xpath('//span[@class="a-size-medium a-color-success"]')
        avail.append(availability.text)
    except NoSuchElementException:
        avail.append('-')

#     print(avail[k])
#     k+=1

for j in range(len(brand_name)):
    if 'Visit the ' in brand_name[j]:
        brand_name[j] = brand_name[j].replace('Visit the ', '').replace(' Store', '')
        
    elif 'Brand: ' in brand_name[j]:
        brand_name[j] = brand_name[j].replace('Brand: ', '')

driver.close()

# creating dataframe to store & visualize the data in a tabular format
PROCUCTS = pd.DataFrame(list(zip(brand_name, product_name, price, ret, exp_del, avail, prod_url)),
                              columns =['BRAND', 'PRODUCT_NAME', 'PRICE', 'RETURN/EXCHANGE', 'EXPECTED_DELIVERY', 'AVAILABILITY', 'PRODUCT_LINK'])


# Saving search data into csv
PROCUCTS.to_csv('guitar.csv')


# displaying the Dataframe
PROCUCTS
```

<h2><font color="red"> Q3: Write a python program to access the search bar and search button on images.google.com and scrape 10 images each for keywords ‘fruits’, ‘cars’, ‘Machine Learning’, ‘Guitar’, ‘Cakes’.</font></h2>

```python
#connecting to google chrome web driver
driver = webdriver.Chrome(r'chromedriver.exe')

#list of search items
to_search = ['fruits', 'cars', 'Machine Learning', 'Guitar', 'Cakes']

#list to store img url
fruits_url = []
cars_url = []
ml_url = []
guitar_url = []
cakes_url = []


# looping through list to get results
for item in to_search:
    # getting the webpage
    driver.get('https://images.google.com/')
    
    
    driver.find_element_by_xpath('//input[@class="gLFyf gsfi"]').send_keys(item)
    driver.find_element_by_class_name('zgAlFc').click()
    
    
    k = 0
    imgResults = driver.find_elements(By.XPATH,"//img[contains(@class,'Q4LuWd')]")
    
    for img in imgResults:        
        if item == 'fruits':
            fruits_url.append(img.get_attribute('src'))
        elif item == 'cars':
            cars_url.append(img.get_attribute('src'))
        elif item == 'Machine Learning':
            ml_url.append(img.get_attribute('src'))
        elif item == 'Guitar':
            guitar_url.append(img.get_attribute('src'))
        elif item == 'Cakes':
            cakes_url.append(img.get_attribute('src'))
        k+=1
        if k==10:
            continue

for i in range(10):
    urllib.request.urlretrieve(str(fruits_url[i]),"web_img/fruits{}.jpg".format(i))
    urllib.request.urlretrieve(str(cars_url[i]),"web_img/cars{}.jpg".format(i))
    urllib.request.urlretrieve(str(ml_url[i]),"web_img/ml{}.jpg".format(i))
    urllib.request.urlretrieve(str(guitar_url[i]),"web_img/guitar{}.jpg".format(i))
    urllib.request.urlretrieve(str(cakes_url[i]),"web_img/cakes{}.jpg".format(i))
    

driver.close()
```

```python
#connecting to google chrome web driver & Maximize current window
options = webdriver.ChromeOptions()
options.add_argument("--start-maximized")
options.add_argument("disable-popup-blocking")
driver = webdriver.Chrome(r'chromedriver.exe', chrome_options=options)

time.sleep(1)


# getting the webpage
driver.get('https://trak.in/')

#waiting to load the web page
time.sleep(2)



#finding required element from the webpage
driver.find_element(By.XPATH,'/html/body/div[1]/header/div[2]/div/div/div/div/nav/ul/li[9]/a').click()


dates = ['March, 2021', 'February, 2021', 'January, 2021']
tables = driver.find_elements(By.XPATH,"//h2[contains(@class,'tablepress')]")

tab_list = []
for tab in tables:
    if tab.text in dates:
        tab_list.append(tab.get_attribute('id').replace('-name',''))
        
#creating empty list to store the data
fund_date = []
startup_name = []
vertical = []
sub_vert = []
city = []
investor = []
inv_type = []
amount = []


for i in tab_list:
    
    x1 = [x.text for x in driver.find_elements(By.XPATH, f'//table[@id = "{i}"]/tbody/tr/td[2]')]
    x2 = [x.text for x in driver.find_elements(By.XPATH, f'//table[@id = "{i}"]/tbody/tr/td[3]')]
    x3 = [x.text for x in driver.find_elements(By.XPATH, f'//table[@id = "{i}"]/tbody/tr/td[4]')]
    x4 = [x.text for x in driver.find_elements(By.XPATH, f'//table[@id = "{i}"]/tbody/tr/td[5]')]
    x5 = [x.text for x in driver.find_elements(By.XPATH, f'//table[@id = "{i}"]/tbody/tr/td[6]')]
    x6 = [x.text for x in driver.find_elements(By.XPATH, f'//table[@id = "{i}"]/tbody/tr/td[7]')]
    x7 = [x.text for x in driver.find_elements(By.XPATH, f'//table[@id = "{i}"]/tbody/tr/td[8]')]
    x8 = [x.text for x in driver.find_elements(By.XPATH, f'//table[@id = "{i}"]/tbody/tr/td[9]')]
    fund_date = fund_date + x1
    startup_name = startup_name + x2
    vertical = vertical + x3
    sub_vert = sub_vert + x4
    city = city + x5
    investor = investor + x6
    inv_type = inv_type + x7
    amount = amount + x8
    

driver.close()


# creating dataframe to store & visualize the data in a tabular format
Funding_Deals = pd.DataFrame({'Date': fund_date, 'Startup Name':startup_name, 'Industry/Vertical':vertical,
                 'Sub-Vertical':sub_vert, 'Location': city, 'Investor':investor,
                 'Investment Type':inv_type, 'Amount(in USD)':amount})

Funding_Deals
```

```python
#connecting to google chrome web driver & Maximize current window
options = webdriver.ChromeOptions()
options.add_argument("--start-maximized")
options.add_argument("disable-popup-blocking")
driver = webdriver.Chrome(r'chromedriver.exe', chrome_options=options)


# getting the webpage
driver.get("https://www.hostelworld.com")
time.sleep(3)

driver.find_element_by_xpath('//input[@class="location-text"]').click()
time.sleep(3)

x = driver.find_element_by_xpath('//input[@class="search-input"]')
x.send_keys("London")
time.sleep(6)

driver.find_element_by_xpath('//div[@class="label"]').click()
time.sleep(3)

driver.find_element_by_xpath('//button[@class="button primary large"]').click()
time.sleep(3)
```

```python
#creating empty list to store the data
hostel_name = []
distance = []
pvt_prices = []
dorms_price = []
rating = []
reviews = []
over_all = []
facilities = []
description =[]
product_url = []

pages = driver.find_elements_by_xpath("//div[@class = 'pagination-item pagination-current' or @class='pagination-item']")
    
for pager in range(len(pages)):
    time.sleep(3)
    
    #hostel name
    try:
        name = driver.find_elements_by_xpath("//h2[@class='title title-6']")
        for i in name:
            hostel_name.append(i.text)
    except NoSuchElementException:
        hostel_name.append('-')
        
    #distance from city centre
    try:
        dist = driver.find_elements_by_xpath("//div[@class='subtitle body-3']/a/span[1]")
        for i in dist:
            distance.append(i.text.replace('Hostel - ','').replace(' from city centre',''))
    except NoSuchElementException:
        distance.append('-')
        
    #privates from price & dorms from price
    for i in driver.find_elements_by_xpath("//div[@class='prices-col']"):
        try:
            pvt_price = driver.find_element_by_xpath("//a[@class='prices']/div[1]/div")
            pvt_prices.append(pvt_price.text)
        except NoSuchElementException:
            pvt_prices.append('-')
        try:
            dorms = driver.find_element_by_xpath("//a[@class='prices']/div[2]/div")
            dorms_price.append(dorms.text)
        except NoSuchElementException:
            dorms_price.append('-')
            
    
    #facilities
    fac = driver.find_elements_by_xpath('//div[@class="facilities-label facilities"]')
    for i in fac:
        try:
            facilities.append(i.text.split('\n'))
        except NoSuchElementException:
            facilities.append('-')
    
    
    #url of each hostel
    p_url = driver.find_elements_by_xpath("//div[@class='prices-col']//a[2]")
    for i in p_url:
        product_url.append(i.get_attribute('href'))
    
    if pager == 0:
        driver.execute_script("window.scrollBy(0,100000)")
        time.sleep(2)
        pages[1].click()




for i in product_url:
    driver.get(i)
    time.sleep(3)
    
    
    #fetching ratings
    try:
        rat = driver.find_element_by_xpath("//div[@class='score orange big' or @class='score gray big']")
        rating.append(rat.text)
    except NoSuchElementException:
        rating.append('-')
    
    
    #fetching total reviews  
    try:
        rws = driver.find_element_by_xpath("//div[@class='reviews']")
        reviews.append(rws.text.replace(' Total Reviews',''))
    except NoSuchElementException:
        reviews.append('-')
    
    
    #fetch overall review
    try:
        overall_rw = driver.find_element_by_xpath("//div[@class='keyword']/span")
        over_all.append(overall_rw.text)
    except NoSuchElementException:
        over_all.append('-')
    
    
    #fetch property description 
    try:
        disc = driver.find_element_by_xpath("//div[@class='content collapse-content']")
        description.append(disc.text)
    except NoSuchElementException:
        description.append('-')
```

```python
# creating dataframe to store & visualize the data in a tabular format
Hostel_Data  = pd.DataFrame({'Hostel_Name':hostel_name, "Distance_from_City_center":distance,'Ratings':rating, 'Total_Review':reviews,'Overall_Reviews':over_all,'Facilities':facilities,"Private_prices":pvt_prices,"dorm_prices":dorms_price,"Property_Description":description})

Hostel_Data.head(20)
```
